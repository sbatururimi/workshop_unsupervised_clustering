## Brown clustering

```python
articles = []
for filename in glob.glob("datasets/*.txt"):
    with open(filename, "r") as f:
        text = f.read()
        articles.append(text)
```

```python
len(articles)
```

```python
data = {"articles":articles}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
sample_data = df["articles"][:100].astype("str").tolist()
sample_data
```

```python
tokenizer = RegexpTokenizer(r"\w+")
sample_data_tokenized = [w.lower() for w in sample_data]
sample_data_tokenized = [tokenizer.tokenize(i) for i in sample_data_tokenized]
```

```python
sample_data_tokenized
```

```python
corpus = Corpus(sample_data_tokenized, 0.001)
clustering = BrownClustering(corpus, 6)
clustering.train()
```

```python
clustering.get_similar("trade")
```

```python
clustering.get_similar("nasdaq")
```

## Keywords extractions

```python
simple_kwextractor = yake.KeywordExtractor()
```

```python
len(df)
```

```python
sentences = []

samples_articles = df["articles"][:100].astype(str)
for article in samples_articles:
    articles_keywords = simple_kwextractor.extract_keywords(article)
    
    sentence_output = ""
    for word, number in articles_keywords:
        sentence_output += word + " "
        
    sentences.append(sentence_output)
```

```python
samples_articles[0]
```

```python
sentences[0]
```

```python
tokenizer = RegexpTokenizer(r"\w+")
sample_data_tokenized = [w.lower() for w in sentences]
sample_data_tokenized = [tokenizer.tokenize(i) for i in sample_data_tokenized]
```

```python
corpus = Corpus(sample_data_tokenized, 0.001)
clustering = BrownClustering(corpus, 6)
clustering.train()
```

```python
clustering.get_similar("nasdaq")
```

```python
clustering.get_similar("trade")
```

```python
len(clustering._codes["trade"])
```

```python
megacluster = clustering.helper.get_cluster(0)

cluster_list = [[]]
list_index = 0

for i in range(len(megacluster) - 1):
    if megacluster[i - 1] < megacluster[i]:
        cluster_list[list_index].append(megacluster[i])
    else:
        # new sublist
        cluster_list.append([])
        list_index += 1
        
        cluster_list[list_index].append(megacluster[i])
```
